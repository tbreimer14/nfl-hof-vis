```python
# load pandas df
import pandas as pd

reg_season = pd.read_csv('all_stats_reg_season_filtered.csv')
mvps = pd.read_csv('MVPs.csv')
playoffs = pd.read_csv('playoff_stats.csv')
sb_wins = pd.read_csv('SB_wins.csv')

# join reg_season & mvps on Player col
all_stats = pd.merge(reg_season, mvps, on='Player', how='left')

# fill NaN values in MVP column with 0
all_stats['MVP'] = all_stats['MVP'].fillna(0)

# join reg_season_mvps & sb_wins on Player col
all_stats = pd.merge(all_stats, sb_wins, on='Player', how='left')

# fill NaN values in SB_Wins column with 0
all_stats['SB_Wins'] = all_stats['SB_Wins'].fillna(0)

# join reg_season_mvps & playoffs on Player col
all_stats = pd.merge(all_stats, playoffs, on='Player', how='left')

display(all_stats)

# export all_stats to csv
all_stats.to_csv('all_stats.csv', index=False)
```
